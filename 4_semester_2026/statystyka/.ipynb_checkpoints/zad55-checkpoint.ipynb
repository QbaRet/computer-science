```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
X=np.random.normal(0,1,100)
e=np.random.normal(0,0.5,100)
Y=-1+0.5*X+e
print("Theta0:", -1, "Theta1:", 0.5)
print("Dlugosc wektora Y:", len(Y))




```

```python
plt.scatter(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Wykres rozrzutu")
plt.show()
```

```python
X_reshaped = X.reshape(-1, 1)


model = LinearRegression()
model.fit(X_reshaped, Y)

beta0_hat = model.intercept_
beta1_hat = model.coef_[0]

print(f"Estymowana Beta0 (Intercept): {beta0_hat:.4f}")
print(f"Estymowana Beta1 (Slope): {beta1_hat:.4f}")

plt.scatter(X, Y)
plt.plot(X, model.predict(X_reshaped), color='red', label='Linia regresji')
plt.legend()
plt.show()
```

```python

x_pop = np.linspace(X.min(), X.max(), 100)
y_pop = -1 + 0.5 * x_pop

plt.scatter(X, Y, alpha=0.5, label='Dane')

plt.plot(X, model.predict(X_reshaped), color='red', label='Linia estymowana (OLS)')
plt.plot(x_pop, y_pop, color='black', linestyle='--', label='Linia populacyjna (Prawda)')

plt.legend()
plt.title("Porównanie modelu z rzeczywistością")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X_reshaped)

model_poly = LinearRegression()
model_poly.fit(X_poly, Y)

print(f"Beta0 (intercept): {model_poly.intercept_:.4f}")
print(f"Beta1 (przy x): {model_poly.coef_[0]:.4f}")
print(f"Beta2 (przy x^2): {model_poly.coef_[1]:.4f}")
```
