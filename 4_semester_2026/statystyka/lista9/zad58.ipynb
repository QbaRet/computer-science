```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

df = pd.read_csv("Auto.csv", na_values='?')
df = df.dropna() 

df['origin'] = df['origin'].astype('category')


train, test = train_test_split(df, test_size=0.2, random_state=42)
model_num = smf.ols('mpg ~ year', data=train).fit()

model_cat = smf.ols('mpg ~ C(year)', data=train).fit()



print(f"Model LICZBOWY - R2: {model_num.rsquared:.4f}")
print(f"Model KATEGORYCZNY - R2: {model_cat.rsquared:.4f}")
print("-" * 40)

pred_num = model_num.predict(test)
pred_cat = model_cat.predict(test)

rmse_num = np.sqrt(mean_squared_error(test['mpg'], pred_num))
rmse_cat = np.sqrt(mean_squared_error(test['mpg'], pred_cat))


print(f"Model LICZBOWY - RMSE: {rmse_num:.4f}")
print(f"Model KATEGORYCZNY - RMSE: {rmse_cat:.4f}")


print("\n--- SZCZEGÓŁY MODELU KATEGORYCZNEGO ---")
print(model_cat.summary().tables[1])
```
