```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split

df = pd.read_csv("Auto.csv", na_values='?')
df = df.dropna() 

df['origin'] = df['origin'].astype('category')


train, test = train_test_split(df, test_size=0.2, random_state=42)

print(train)
```

```python
model_a = smf.ols('mpg ~ horsepower', data=train).fit()
print(model_a.summary())
```

```python
model_b = smf.ols('mpg ~ horsepower + weight + cylinders', data=train).fit()
print(model_b.summary())
```

```python
model_c=smf.ols('mpg ~ np.log(horsepower)', data=train).fit()
print(model_c.summary())
```

```python
model_d=smf.ols('mpg ~ horsepower * weight', data=train).fit()
print(model_d.summary())
```
