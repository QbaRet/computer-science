```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
n = 100000  
tablica = [i for i in range(1, 21)]
objetosci_mc = []
objetosci_dokladne=[]
for d in tablica:
    
    x = np.random.uniform(-1, 1, (n, d))
    odleglosci=np.linalg.norm(x,axis=1)
    trafione=np.sum(odleglosci<=1)
    
    Vhiper=2**d
    Vkuli=(trafione*Vhiper) /n
    objetosci_mc.append(Vkuli)

    Vkuli_dokladne = (np.pi**(d/2)) / gamma(d/2 + 1)
    objetosci_dokladne.append(Vkuli_dokladne)
    
plt.plot(tablica, objetosci_mc, marker='o', label='Symulacja Monte Carlo')
plt.plot(tablica, objetosci_dokladne, marker='x', linestyle='--', color='red', label='Wzór Dokładny')
plt.xlabel('Liczba wymiarów d')
plt.ylabel('Objętość hiperkuli (V_d)')
plt.title('Paradoks objętości: Hiperkula vs Hiperkostka')
plt.legend()
plt.grid(True)
plt.show()
```
